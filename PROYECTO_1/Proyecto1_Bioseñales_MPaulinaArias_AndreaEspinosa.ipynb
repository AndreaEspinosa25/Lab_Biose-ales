**Enfermedad de Huntington**

La enfermedad de Huntington es una afección hereditaria caracterizada por la degeneración progresiva de las células nerviosas en ciertas áreas del cerebro. Esta enfermedad provoca movimientos incontrolados, deterioro cognitivo y problemas emocionales. Los síntomas generalmente aparecen en la edad adulta, entre los 30 y 50 años, aunque en algunos casos pueden manifestarse en la juventud. La enfermedad se debe a mutaciones en el gen HTT y se hereda de manera autosómica dominante (significa que la afección genética ocurre cuando el niño hereda una sola copia de un gen mutado (cambiado) de uno de los padres) [2].

Con el fin de establecer un diagnóstico preciso de la enfermedad de Huntington (SH), hay que basarse en los síntomas y signos clínicos de una persona que presenta confirmación de herencia autosómica dominante por parte de uno de sus progenitores. Inicialmente, es crucial obtener una historia clínica que incluya antecedentes patológicos familiares. El estándar actual consiste en la determinación del ADN, que evidencie una repetición de CAG de al menos 36 en el gen Huntingtina, localizado en el cromosoma 4 [2].

Además, los estudios de imágenes, como la tomografía computarizada y la resonancia magnética, revelan una disminución gradual de los ganglios basales y una atrofia en las cortezas frontal y temporal. En el caso del diagnóstico prenatal, se puede llevar a cabo entre las semanas 10 y 12 de embarazo mediante una muestra de las vellosidades coriónicas, y entre las semanas 15 y 17 mediante amniocentesis. Este tipo de diagnóstico se recomienda especialmente a los padres que conocen su estado genético. En situaciones donde se confirme la presencia del gen Huntingtina en el embrión, se plantea la posibilidad de interrumpir el embarazo. En el proceso de diagnóstico diferencial, es esencial comparar los síntomas motores con otras patologías y establecer relaciones para discernir entre distintas enfermedades [2].

Esta enfermedad presenta dificultades en su diagnóstico debido a que en la gran mayoría de los casos, la enfermedad se desarrolla en la edad adulta y suele ser de progresión lenta, los primeros síntomas pueden pasar inadvertidos para familiares y pacientes a lo largo de los años confundiéndose con problemas de aprendizaje, lo que retrasaría el tratamiento de la enfermedad (que solo es para mejorar la calidad de vida de la persona) [3].


**Esclerosis Lateral**

La esclerosis lateral amiotrófica (ELA), también llamada la enfermedad de Lou Gehrig, es un trastorno neurológico, degenerativo y mortal que afecta las neuronas motoras ubicadas en la corteza frontal, el tronco encefálico y la médula espinal. La enfermedad tiene una amplia gama de presentación, generalmente comienza como debilidad muscular en una extremidad, cambios en el timbre de la voz o dificultad para deglutir, que progresa a debilidad generalizada y parálisis de los músculos respiratorios que conducen a la muerte por insuficiencia respiratoria [4].

La ELA se diagnostica clínicamente. Las pruebas complementarias contribuyen a confirmar la sospecha diagnóstica y descartar otras patologías. Actualmente, establecer la certeza diagnóstica requiere la participación de ambas neuronas motoras, o la participación de las neuronas motoras inferiores en al menos dos de cuatro regiones (bulbar, cervical, torácica o lumbosacra), así como la exclusión de otras condiciones etiológicas [4].

Un diagnóstico definitivo aún requiere una historia clínica, un examen neurológico y pruebas complementarias. La prueba más importante es el electromiograma (EMG), que revela una disminución leve en la conducción motora, con amplitud de potencial de acción reducida y denervación muscular aguda o en curso, como la presencia potencial de fibrilación y fasciculaciones en múltiples músculos en las regiones examinadas. Aunque los hallazgos en el EMG permiten distinguir las atrofias neurogénicas de las enfermedades musculares, algunos procesos miopáticos pueden producir confusión con la ELA, como la polimiositis crónica, la enfermedad de Pompe o la neuropatía motora multifocal con bloqueos de conducción. Tradicionalmente, las pruebas genéticas no han sido una parte rutinaria de la evaluación de la ELA. Si hay antecedentes familiares, se podrían considerar las pruebas genéticas para el diagnóstico presintomático [4].

El diagnóstico de la ELA puede presentar dificultades por que esta puede ser mal diagnosticada y confundida con otras enfermedades neuromusculares debido a los sintomas que presenta, además aún no existe ninguna prueba patognomónica (prueba que permita obtener un hallazgos distintivo o característico de la enfermedad o afección y que se pueden usar para establecer un diagnóstico) que facilite el reconocimiento de esta [4].


**Detrend - Scipy**

La función detrend de scipy permite eliminar las tendencias lineales de los datos, lo que permite que la solo quede la parte oscilatoria de la señal, haciendo que el análisis de la señal sea fácil de realizar. La eliminación de esta tendencia se puede interpretar como la resta de un polinomio de ajustes de mínimos cuadrados [5].

Desde una vista matemática se tiene que [6]:
Una serie temporal se puede descomponer en tres componentes principales:
- X(t): la serie observada en el tiempo t.
- T(t): la tendencia.
- S(t): los patrones estacionales o cíclicos.
- R(t): el ruido o las fluctuaciones residuales (lo que queda una vez eliminada la tendencia y la estacionalidad).

Esto se puede representar así [6]:

X(t) = T(t) + S(t) + R(t)

El objetivo es eliminar T(t), que es la tendencia, dejando S(t) + R(t).
Luego, si se asume que la tendencia es una constante, es decir, simplemente la media de los datos. La fórmula básica sería [6]:

Xdetrended(t) = X(t) - µ

donde µ es la media de la serie:

µ = 1/N(∑_(i=1)^n X(i))

se resta la media de la serie temporal a cada valor de la serie. Esto desplaza los datos para que estén centrados alrededor de cero, eliminando una tendencia constante o estacionaria [6].

Se asume que la tendencia sigue una línea recta, de la forma [6]:

T(t) = a*t + b

Donde [6]:

a = es la pendiente de la tendencia lineal

b = es la intersección en el eje y el valor inicial de la tendencia.

Para eliminar la tendencia, se ajusta una línea recta a los datos. El proceso es buscar los coeficientes a y b que minimicen el error entre la serie observada X(t) y esta tendencia lineal [6].

El error se minimiza a través del método de mínimos cuadrados:

min∑_(X(t)- (a*t + b))^2

Una vez obtenemos a y b, restamos la tendencia de los datos:

Xdetrended(t) = X(t) - (a*t + b)

Este proceso remueve la tendencia lineal, dejando las fluctuaciones que ocurren alrededor de esta línea [6].



**El Filtro Hampel**

Es un filtro no lineal de limpieza de datos que busca los valores atípicos o datos anormales locales en una secuencia temporal. Los valores atípicos son muy variados y por lo general son una “complicación importante” en el análisis de datos, razón por la cual es crucial considerar su origen; los errores de muestreo y los errores ajenos al muestreo (errores de reporte, errores sin respuesta y errores de selección de la muestra) son ejemplos de ello [7].

El filtro Hampel busca los datos anormales locales en una secuencia temporal, estos valores son sustituidos por valores alternativos más razonables una vez encontrados; el algoritmo usa la Desviación de la Mediana Absoluta (MAD) en Series Temporales, es decir, utiliza la mediana de las observaciones vecinas como valor de referencia y el estimador de escala de MAD como medida alternativa de la distancia, en otras palabras, un punto de datos es declarado como valor atípico y puede ser sustituido si se encuentra a más de un número t de la escala de MAD estimada de la mediana respecto de sus vecinos, el valor de reemplazo utilizado en este procedimiento es la Mediana. Más específicamente, para cada observación en la secuencia de datos original; el filtro Hampel construye una ventana “móvil” que incluye los puntos K anteriores, el punto de datos de interés primario, y los puntos K subsiguientes de datos. El valor de referencia utilizado para el punto central de datos es la Mediana de estos 2k+1 observaciones sucesivas, y la estimación escala MAD se calcula a partir de estas mismas observaciones para servir como una medida de la "propagación natural local" de la secuencia de datos. Si el punto central de datos se encuentra a más de t valores de la estimación de la escala MAD de la mediana, es reemplazado por la Mediana, de lo contrario, se deja sin cambios [7].

**formulación matematica [8]**

Dada una secuencia x1, x2, x3, …, xn y una ventana deslizante de longitud k, defina las estimaciones de la mediana y la desviación estándar punto a punto mediante:

Mediana local

—mi=median(xi−k,xi−k+1,xi−k+2,…,xi,…,xi+k−2,xi+k−1,xi+k)

Desviación estándar

— σi=κmedian(∣xi−k−mi∣,…,∣xi+k−mi∣),

donde κ=1/√2erf−1(1/2)≈1.4826

La cantidad σi /κ se conoce como desviación absoluta media (MAD).

Si una muestra xi es tal que

∣xi−mi∣>nσσi

para un umbral dado nσ, el identificador de Hampel declara xi como un valor atípico y lo sustituye por mi.



**Entropía de permutación**

La entropía de permutación es una medida de complejidad para series temporales y mide el grado de desorden en una serie temporal comparando el orden relativo de valores consecutivos. Esto se hace observando cómo cambian los valores vecinos en una serie de datos y determinando todas las permutaciones posibles para un conjunto de datos de una longitud específica (n).
Esto tiene la ventaja de dar resultados significativos en presencia de ruido observacional y dinámico, además, es un método simple, de cálculo rápido y robusto. Para una serie de tiempo {𝑥𝑡 } = 1, . . . , 𝑇 , 𝜋 permutaciones de orden 𝑛, la PE se define por [9]:

𝐻(𝑛) = − ∑ 𝑝(𝜋)𝑙𝑜𝑔(𝑝(𝜋))

Donde P(𝜋) es la frecuencia relativa de cada permutación 𝜋 dentro de la serie temporal. Este valor representa la cantidad de información contenida en la comparación de 𝑛 valores consecutivos de la serie. Un sistema completamente ordenado tendría una entropía de permutación baja, mientras que un sistema altamente caótico o aleatorio tendría una entropía más alta [9].

En el ámbito de la bioingeniería es muy usada para el análisis de señales obtenidas por pruebas de ECG, EMG y EEG ya que permite obtener irregularidades en la dinámica de la señal, otro aplicativo es el uso de la EP para el estudio de las sub-bandas de señales obtenidas por medio de magnetoencefalografía (MEG) y encontrar características que sirven para la detección del deterioro cognitivo leve (MCI), también es usado para la caracterización de texturas en imágenes digitales. Hay estudios que muestran el uso de la EP para el estudio del efecto de la anestesia (efecto de un fármaco anestésico en las grabaciones de EEG al usar anestesia con sevoflurano). La EP es usada en diferentes áreas y tiene variedad de aplicativos incluso en el mercado económicos [10].

# Hampel Filter

```python
def hampel_filter(wk, K, FilterParams):

    Thresh = FilterParams[0] #Definir umbral de cambio

    ref = np.median(wk) #Calcular la mediana de la ventana wk

    #Calcular la desv absoluta media (MAD)
    AbsDev = np.abs(wk - ref)
    MAD = 1.4826 * np.median(AbsDev)  #Factor de escala para distribución normal

    #Valor de prueba para identificar outliers
    TestVal = np.abs(wk - ref)

    # Copia de wk para no modificar los datos originales
    yk = np.copy(wk)

    #Número de elementos que superan el umbral
    outliers = TestVal > (Thresh * MAD)
    numero = np.sum(outliers)  #Cantidad de outliers detectados
    #print("Número de outliers detectados:", numero)

    # Reemplazar los outliers por la mediana
    yk[outliers] = ref

    return yk
```

# Coeficiente de variación y desviación estándar

```python
def funcionCVSD(data):
  SD = np.std(data)  # ddof=1 para muestra (desviación estándar muestral)
  CV = 100*(SD/np.mean(data)) if np.mean(data) != 0 else np.nan
  return CV, SD
```

# Entropía de permutación

```python
#Delay entropia de permutación
def DataDelay(data, delay=1, dimension=3):
    if delay*dimension > len(data):
        raise NameError('No es posible')
    NuevoData = np.array([data[0:len(data)-delay*(dimension - 1)]])
    for i in range(1, dimension):
        NuevoData = np.append(NuevoData, [data[i*delay:len(data) - delay*(dimension - i - 1)]], axis=0)
    return NuevoData;
a = np.arange(0,3)
perms = set(permutations(a))
```

```python
# Entropia de permutación
def permutacionEntropia(rank):
  PEd = 0
  for rank_pattern in perms:
    copy_rank = np.tile(rank_pattern,(rank[0].size,1))
    resultado = copy_rank.transpose() -rank
    total_pattern = np.shape(sum(np.abs(resultado),0))[0] - np.count_nonzero(sum(np.abs(resultado),0))
    if total_pattern != 0:
      PEd = PEd + (total_pattern/230)*np.log2(total_pattern/230)
  return PEd
```

La marcha, al ser un proceso motor controlado por el sistema nervioso central, puede verse alterada en personas con estas patologías [11], lo que hace que medir la variabilidad en la marcha ofrezca información valiosa sobre la severidad y naturaleza del deterioro motor en cada enfermedad.

Para realizar esta comparación, se utilizaron tres métricas clave en las fases de paso (Stride) y balanceo (Swing), tanto en el lado izquierdo como en el derecho:


*   Entropía (Ent): Evalúa el grado de desorden o imprevisibilidad en los patrones de marcha. Una entropía más alta refleja mayor desorganización o variabilidad [12].

*   Coeficiente de Variación (CV): Mide la variabilidad relativa (normalizada) en los ciclos de marcha [13].

* Desviación Estándar (SD): Mide la dispersión o variabilidad absoluta en los ciclos de marcha [13].

Para facilitar la identificación de estos patrones, se realizan diferentes tipos de diagramas y pruebas de hipótesis.

#Pruebas de hipótesis

El análisis de las pruebas de hipótesis permite tomas decisiones estadísticas basadas en datos, con el fin de determinar si los resultados observados son significativos.

La prueba de normalidad de realizó mediante la **prueba de Shapiro-Wilks** debido a que se trabajaron grupos caracterizados por tener menos de 50 muestras [17].

Seguidamente, se realizó la **prueba de homocedasticidad de Levene** para verificar si varios grupos de datos tienen varianzas iguales. Esta prueba es más robusta que otras pruebas de homocedasticidad (como la prueba de Bartlett) cuando los datos no siguen una distribución normal, lo que la hace adecuada para este análisis [18].

Por último, como no se cumple el supuesto de normalidad para ningún caso, y no se cumplen los de homocedasticidad en muchos casos, se opta por la realización de pruebas no paramétricas, en este caso, se emplea la **Prueba U de Mann-Whitney** para comparar si hay diferencias significativas entre las distribuciones de dos grupos independientes.  A diferencia de pruebas paramétricas como el t-test, la prueba de Mann-Whitney no requiere que los datos sigan una distribución normal, lo que la hace adecuada para este análisis [19].

```python
#Se crea una función que permita la comparación de los diferentes grupos
def comparar_grupos(data1, data2, label1, label2, columna, normalidad_cumplida, homocedasticidad_cumplida):
    alpha = 0.05
    results = {}

    if normalidad_cumplida and homocedasticidad_cumplida:
        t_stat, t_p_value = stats.ttest_ind(data1, data2)
        results['t_stat'] = t_stat
        results['p_value'] = t_p_value
        results['conclusion'] = (t_p_value < alpha)
    else:
        #Ho = No hay diferencias significativas entre los grupos (pertecenen a una misma población)
        #Ha = Hay diferencias significativas entre los grupos (No pertenecen a una misma población)

        u_stat, u_p_value = stats.mannwhitneyu(data1, data2)
        results['u_stat'] = u_stat
        results['p_value'] = u_p_value
        results['conclusion'] = (u_p_value < alpha)

    results['Grupo 1'] = label1
    results['Grupo 2'] = label2
    return results

#Se crean diccionarios vacios donde se guardaran los datos obtenidos
resultados_normalidad = {}
resultados_homocedasticidad = {}
resultados_comparaciones = {}

#Se realiza un for que permita hacer cada una de las pruebas por cada columna
for columna in df.columns[2:]:


    groups = {condition: df[df['Condicion'] == condition][columna] for condition in df['Condicion'].unique()}

    # Prueba de normalidad

    #Ho= Los datos siguen una distribución normal
    #Ha= Los datos NO siguen una distribución normal
    normality_results = {condition: stats.shapiro(data)[1] for condition, data in groups.items()}
    normalidad_cumplida = all(p > 0.05 for p in normality_results.values())
    resultados_normalidad[columna] = 'Dist normal' if normalidad_cumplida else 'NO dist normal'

    # Prueba de homocedasticidad

    #Ho= La varianza es igual entre diferentes grupos (homocedasticidad)
    #Ha= La varianza NO es igual en los diferentes grupos (heterocedasticidad)
    p_levene = stats.levene(*groups.values())[1]
    homocedasticidad_cumplida = p_levene > 0.05
    resultados_homocedasticidad[columna] = 'Hay homocedasticidad' if homocedasticidad_cumplida else 'Hay heterocedasticidad'

    #Se compara los resultados entre los diferentes grupos
    comparaciones = []
    for i in range(len(groups)):
        for j in range(i + 1, len(groups)):
            label1, label2 = list(groups.keys())[i], list(groups.keys())[j]
            resultados = comparar_grupos(groups[label1], groups[label2], label1, label2, columna, normalidad_cumplida, homocedasticidad_cumplida)
            comparaciones.append(resultados)

    #Se crea un dataFrame de los datos obtenidos de las diferentes comparaciones
    resultados_comparaciones[columna] = pd.DataFrame(comparaciones)

#Se convierten estos resultados en dataFrame
df_normalidad = pd.DataFrame.from_dict(resultados_normalidad, orient='index', columns=['Conclusión'])
df_homocedasticidad = pd.DataFrame.from_dict(resultados_homocedasticidad, orient='index', columns=['Conclusión'])
df_comparaciones = {columna: df for columna, df in resultados_comparaciones.items()}

#Se crea un for que permita reemplazar el valor booleano por una string que indique la conclusión o resultado de la comparación analizada
for columna, df in df_comparaciones.items():
    df['conclusion'] = df['conclusion'].replace({True: 'Dif SIGNFICATIVA', False: 'NO significativa'})

#Se muestran los resultados
print("\nResultados de Normalidad:")
print(df_normalidad)

print("\nResultados de Homocedasticidad:")
print(df_homocedasticidad)

for columna, df in df_comparaciones.items():
    print(f"\nResultados de Comparaciones (Prueba U) para {columna}:")
    print(df[['Grupo 1', 'Grupo 2', 'p_value', 'conclusion']])
```

### **1. Normalidad:**

Los resultados muestran que ninguna de las variables evaluadas sigue una distribución normal. Esto implica que los métodos estadísticos paramétricos tradicionales no son apropiados para realizar comparaciones entre los grupos. En su lugar, se utilizan pruebas no paramétricas, como la prueba U de Mann-Whitney.

### **2. Homocedasticidad:**

En el artículo base, ALS presentaba un aumento en la variabilidad respecto a los controles, pero menor que en Huntington, lo que concuerda con las observaciones del análisis realizado, confirmando que la SD es un buen indicador de la variabilidad en la marcha. Huntington tiene la mayor SD, seguido por ALS y luego Parkinson.

**3.	Entropía**

La mayor entropía en ALS, seguida de Huntington y luego Parkinson, está en línea con los aumentos de complejidad y variabilidad observados en el artículo base, aunque este no discute directamente la entropía, reporta un aumento en la complejidad de las fluctuaciones y perturbaciones en la dinámica de la marcha en ALS. # Medidas útiles para discriminar las poblaciones clínicas de los controles

En base a los resultados obtenidos, las medidas más útiles que podrían servir para discriminar entre las poblaciones clínicas (ALS, Huntington y Parkinson) y el grupo Control, son:

1. **Entropía (Ent)**

    La entropía presenta una mayor dispersión en ALS (valores más bajos) en comparación con los controles, lo que sugiere una desorganización significativa en los patrones de marcha. Este hallazgo puede servir como un indicador clave para diferenciar entre estos grupos. Así mismo, se observaron diferencias entre ALS y Huntington o Parkinson. Sin embargo, no hay diferencias significativas entre Huntington y Parkinson, lo que indica que la entropía puede ser más efectiva para discriminar a ALS de los otros grupos.
    
    Por otro lado, heterocedasticidad observada en la entropía sugiere que la variabilidad de marcha entre los grupos es diferente. Esto puede reforzar la idea de que esta medida es relevante para identificar y discriminar entre las poblaciones clínicas y controles.


2. **Desviación estándar (SD)**

    La desviación estándar mostró diferencias significativas entre el grupo Control y los tres grupos clínicos en la fase de paso y balanceo. Los controles presentan valores bajos y consistentes, mientras que los pacientes con ALS, Huntington y Parkinson muestran variabilidad aumentada.

    Aunque ALS, Huntington y Parkinson presentan variabilidad similar en algunas mediciones, la diferencia significativa entre Huntington y Parkinson en SD_RStride sugiere que esta métrica puede ayudar a distinguir entre estos dos grupos en la fase de marcha derecha.


Estas medidas proporcionan un enfoque robusto para identificar diferencias en los patrones de marcha y pueden ser utilizadas para apoyar conclusiones sobre el impacto de las enfermedades neurodegenerativas en la marcha

Es importante aclarar que, el coeficiente de variación parece tener un valor limitado como diferenciador entre el grupo de personas control y con enfermedades. Graficamente, se tiene que, la dispersión visual sugiere diferencias. Sin embargo, los gráficos pueden hacer que las diferencias parezcan más significativas de lo que realmente son en términos estadísticos. Esto se comprueba con las pruebas realizadas, las cuales muestran que dichas diferencias no son estadísticamente significativas, por lo que el CV no se considera un diferenciador robusto entre grupos, salvo en casos específicos como entre Parkinson y Huntington en la fase de marcha.

# Conclusiones


Las alteraciones en la marcha observadas en pacientes con enfermedades neurodegenerativas sugieren que su análisis puede ser una herramienta útil tanto para el diagnóstico temprano como para el seguimiento de la progresión de estas enfermedades. Cada patología estudiada muestra un patrón distintivo en las alteraciones de la marcha, lo que refleja la diversidad de los mecanismos patológicos subyacentes y su impacto en la locomoción, generando una heterogeneidad en los patrones de marcha.

En este contexto, la entropía y la desviación estándar se caracterizan como las métricas más útiles para diferenciar entre las diversas patologías, mientras que el coeficiente de variación (CV) parece no aportar tanta información significativa en el análisis estadísico. En particular, los valores de CV son similares entre los grupos, lo que sugiere que puede no ser la mejor herramienta para detectar variabilidad en los patrones de marcha.

Este análisis refuerza la idea de que las enfermedades neurodegenerativas como ALS, Huntington y Parkinson impactan notablemente la variabilidad y la irregularidad de la marcha, siendo la entropía y el coeficiente de variación mejores indicadores de estas alteraciones.
