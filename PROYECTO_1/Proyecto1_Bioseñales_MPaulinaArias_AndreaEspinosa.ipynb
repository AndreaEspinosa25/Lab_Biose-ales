**Detrend - Scipy**

La funci√≥n detrend de scipy permite eliminar las tendencias lineales de los datos, lo que permite que la solo quede la parte oscilatoria de la se√±al, haciendo que el an√°lisis de la se√±al sea f√°cil de realizar. La eliminaci√≥n de esta tendencia se puede interpretar como la resta de un polinomio de ajustes de m√≠nimos cuadrados [5].

Desde una vista matem√°tica se tiene que [6]:
Una serie temporal se puede descomponer en tres componentes principales:
- X(t): la serie observada en el tiempo t.
- T(t): la tendencia.
- S(t): los patrones estacionales o c√≠clicos.
- R(t): el ruido o las fluctuaciones residuales (lo que queda una vez eliminada la tendencia y la estacionalidad).

Esto se puede representar as√≠ [6]:

X(t) = T(t) + S(t) + R(t)

El objetivo es eliminar T(t), que es la tendencia, dejando S(t) + R(t).
Luego, si se asume que la tendencia es una constante, es decir, simplemente la media de los datos. La f√≥rmula b√°sica ser√≠a [6]:

Xdetrended(t) = X(t) - ¬µ

donde ¬µ es la media de la serie:

¬µ = 1/N(‚àë_(i=1)^n X(i))

se resta la media de la serie temporal a cada valor de la serie. Esto desplaza los datos para que est√©n centrados alrededor de cero, eliminando una tendencia constante o estacionaria [6].

Se asume que la tendencia sigue una l√≠nea recta, de la forma [6]:

T(t) = a*t + b

Donde [6]:

a = es la pendiente de la tendencia lineal

b = es la intersecci√≥n en el eje y el valor inicial de la tendencia.

Para eliminar la tendencia, se ajusta una l√≠nea recta a los datos. El proceso es buscar los coeficientes a y b que minimicen el error entre la serie observada X(t) y esta tendencia lineal [6].

El error se minimiza a trav√©s del m√©todo de m√≠nimos cuadrados:

min‚àë_(X(t)- (a*t + b))^2

Una vez obtenemos a y b, restamos la tendencia de los datos:

Xdetrended(t) = X(t) - (a*t + b)

Este proceso remueve la tendencia lineal, dejando las fluctuaciones que ocurren alrededor de esta l√≠nea [6].



**El Filtro Hampel**

Es un filtro no lineal de limpieza de datos que busca los valores at√≠picos o datos anormales locales en una secuencia temporal. Los valores at√≠picos son muy variados y por lo general son una ‚Äúcomplicaci√≥n importante‚Äù en el an√°lisis de datos, raz√≥n por la cual es crucial considerar su origen; los errores de muestreo y los errores ajenos al muestreo (errores de reporte, errores sin respuesta y errores de selecci√≥n de la muestra) son ejemplos de ello [7].

El filtro Hampel busca los datos anormales locales en una secuencia temporal, estos valores son sustituidos por valores alternativos m√°s razonables una vez encontrados; el algoritmo usa la Desviaci√≥n de la Mediana Absoluta (MAD) en Series Temporales, es decir, utiliza la mediana de las observaciones vecinas como valor de referencia y el estimador de escala de MAD como medida alternativa de la distancia, en otras palabras, un punto de datos es declarado como valor at√≠pico y puede ser sustituido si se encuentra a m√°s de un n√∫mero t de la escala de MAD estimada de la mediana respecto de sus vecinos, el valor de reemplazo utilizado en este procedimiento es la Mediana. M√°s espec√≠ficamente, para cada observaci√≥n en la secuencia de datos original; el filtro Hampel construye una ventana ‚Äúm√≥vil‚Äù que incluye los puntos K anteriores, el punto de datos de inter√©s primario, y los puntos K subsiguientes de datos. El valor de referencia utilizado para el punto central de datos es la Mediana de estos 2k+1 observaciones sucesivas, y la estimaci√≥n escala MAD se calcula a partir de estas mismas observaciones para servir como una medida de la "propagaci√≥n natural local" de la secuencia de datos. Si el punto central de datos se encuentra a m√°s de t valores de la estimaci√≥n de la escala MAD de la mediana, es reemplazado por la Mediana, de lo contrario, se deja sin cambios [7].

**Entrop√≠a de permutaci√≥n**

La entrop√≠a de permutaci√≥n es una medida de complejidad para series temporales y mide el grado de desorden en una serie temporal comparando el orden relativo de valores consecutivos. Esto se hace observando c√≥mo cambian los valores vecinos en una serie de datos y determinando todas las permutaciones posibles para un conjunto de datos de una longitud espec√≠fica (n).
Esto tiene la ventaja de dar resultados significativos en presencia de ruido observacional y din√°mico, adem√°s, es un m√©todo simple, de c√°lculo r√°pido y robusto. Para una serie de tiempo {ùë•ùë° } = 1, . . . , ùëá , ùúã permutaciones de orden ùëõ, la PE se define por [9]:

ùêª(ùëõ) = ‚àí ‚àë ùëù(ùúã)ùëôùëúùëî(ùëù(ùúã))

En el √°mbito de la bioingenier√≠a es muy usada para el an√°lisis de se√±ales obtenidas por pruebas de ECG, EMG y EEG ya que permite obtener irregularidades en la din√°mica de la se√±al, otro aplicativo es el uso de la EP para el estudio de las sub-bandas de se√±ales obtenidas por medio de magnetoencefalograf√≠a (MEG) y encontrar caracter√≠sticas que sirven para la detecci√≥n del deterioro cognitivo leve (MCI), tambi√©n es usado para la caracterizaci√≥n de texturas en im√°genes digitales. # Hampel Filter

```python
def hampel_filter(wk, K, FilterParams):

    Thresh = FilterParams[0] #Definir umbral de cambio

    ref = np.median(wk) #Calcular la mediana de la ventana wk

    #Calcular la desv absoluta media (MAD)
    AbsDev = np.abs(wk - ref)
    MAD = 1.4826 * np.median(AbsDev)  #Factor de escala para distribuci√≥n normal

    #Valor de prueba para identificar outliers
    TestVal = np.abs(wk - ref)

    # Copia de wk para no modificar los datos originales
    yk = np.copy(wk)

    #N√∫mero de elementos que superan el umbral
    outliers = TestVal > (Thresh * MAD)
    numero = np.sum(outliers)  #Cantidad de outliers detectados
    #print("N√∫mero de outliers detectados:", numero)

    # Reemplazar los outliers por la mediana
    yk[outliers] = ref

    return yk
```

```python
# Entropia de permutaci√≥n
def permutacionEntropia(rank):
  PEd = 0
  for rank_pattern in perms:
    copy_rank = np.tile(rank_pattern,(rank[0].size,1))
    resultado = copy_rank.transpose() -rank
    total_pattern = np.shape(sum(np.abs(resultado),0))[0] - np.count_nonzero(sum(np.abs(resultado),0))
    if total_pattern != 0:
      PEd = PEd + (total_pattern/230)*np.log2(total_pattern/230)
  return PEd
```

#Pruebas de hip√≥tesis

El an√°lisis de las pruebas de hip√≥tesis permite tomas decisiones estad√≠sticas basadas en datos, con el fin de determinar si los resultados observados son significativos.

La prueba de normalidad de realiz√≥ mediante la **prueba de Shapiro-Wilks** debido a que se trabajaron grupos caracterizados por tener menos de 50 muestras [17].

Seguidamente, se realiz√≥ la **prueba de homocedasticidad de Levene** para verificar si varios grupos de datos tienen varianzas iguales. Esta prueba es m√°s robusta que otras pruebas de homocedasticidad (como la prueba de Bartlett) cuando los datos no siguen una distribuci√≥n normal, lo que la hace adecuada para este an√°lisis [18].

Por √∫ltimo, como no se cumple el supuesto de normalidad para ning√∫n caso, y no se cumplen los de homocedasticidad en muchos casos, se opta por la realizaci√≥n de pruebas no param√©tricas, en este caso, se emplea la **Prueba U de Mann-Whitney** para comparar si hay diferencias significativas entre las distribuciones de dos grupos independientes.  A diferencia de pruebas param√©tricas como el t-test, la prueba de Mann-Whitney no requiere que los datos sigan una distribuci√≥n normal, lo que la hace adecuada para este an√°lisis [19].

```python
#Se crea una funci√≥n que permita la comparaci√≥n de los diferentes grupos
def comparar_grupos(data1, data2, label1, label2, columna, normalidad_cumplida, homocedasticidad_cumplida):
    alpha = 0.05
    results = {}

    if normalidad_cumplida and homocedasticidad_cumplida:
        t_stat, t_p_value = stats.ttest_ind(data1, data2)
        results['t_stat'] = t_stat
        results['p_value'] = t_p_value
        results['conclusion'] = (t_p_value < alpha)
    else:
        #Ho = No hay diferencias significativas entre los grupos (pertecenen a una misma poblaci√≥n)
        #Ha = Hay diferencias significativas entre los grupos (No pertenecen a una misma poblaci√≥n)

        u_stat, u_p_value = stats.mannwhitneyu(data1, data2)
        results['u_stat'] = u_stat
        results['p_value'] = u_p_value
        results['conclusion'] = (u_p_value < alpha)

    results['Grupo 1'] = label1
    results['Grupo 2'] = label2
    return results

#Se crean diccionarios vacios donde se guardaran los datos obtenidos
resultados_normalidad = {}
resultados_homocedasticidad = {}
resultados_comparaciones = {}

#Se realiza un for que permita hacer cada una de las pruebas por cada columna
for columna in df.columns[2:]:


    groups = {condition: df[df['Condicion'] == condition][columna] for condition in df['Condicion'].unique()}

    # Prueba de normalidad

    #Ho= Los datos siguen una distribuci√≥n normal
    #Ha= Los datos NO siguen una distribuci√≥n normal
    normality_results = {condition: stats.shapiro(data)[1] for condition, data in groups.items()}
    normalidad_cumplida = all(p > 0.05 for p in normality_results.values())
    resultados_normalidad[columna] = 'Dist normal' if normalidad_cumplida else 'NO dist normal'

    # Prueba de homocedasticidad

    #Ho= La varianza es igual entre diferentes grupos (homocedasticidad)
    #Ha= La varianza NO es igual en los diferentes grupos (heterocedasticidad)
    p_levene = stats.levene(*groups.values())[1]
    homocedasticidad_cumplida = p_levene > 0.05
    resultados_homocedasticidad[columna] = 'Hay homocedasticidad' if homocedasticidad_cumplida else 'Hay heterocedasticidad'

    #Se compara los resultados entre los diferentes grupos
    comparaciones = []
    for i in range(len(groups)):
        for j in range(i + 1, len(groups)):
            label1, label2 = list(groups.keys())[i], list(groups.keys())[j]
            resultados = comparar_grupos(groups[label1], groups[label2], label1, label2, columna, normalidad_cumplida, homocedasticidad_cumplida)
            comparaciones.append(resultados)

    #Se crea un dataFrame de los datos obtenidos de las diferentes comparaciones
    resultados_comparaciones[columna] = pd.DataFrame(comparaciones)

#Se convierten estos resultados en dataFrame
df_normalidad = pd.DataFrame.from_dict(resultados_normalidad, orient='index', columns=['Conclusi√≥n'])
df_homocedasticidad = pd.DataFrame.from_dict(resultados_homocedasticidad, orient='index', columns=['Conclusi√≥n'])
df_comparaciones = {columna: df for columna, df in resultados_comparaciones.items()}

#Se crea un for que permita reemplazar el valor booleano por una string que indique la conclusi√≥n o resultado de la comparaci√≥n analizada
for columna, df in df_comparaciones.items():
    df['conclusion'] = df['conclusion'].replace({True: 'Dif SIGNFICATIVA', False: 'NO significativa'})

#Se muestran los resultados
print("\nResultados de Normalidad:")
print(df_normalidad)

print("\nResultados de Homocedasticidad:")
print(df_homocedasticidad)

for columna, df in df_comparaciones.items():
    print(f"\nResultados de Comparaciones (Prueba U) para {columna}:")
    print(df[['Grupo 1', 'Grupo 2', 'p_value', 'conclusion']])
```
