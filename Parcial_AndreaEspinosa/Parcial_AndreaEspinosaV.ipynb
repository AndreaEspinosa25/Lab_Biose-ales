```python
import matplotlib.pyplot as plt
import numpy as np
Fo = 80 # Frecuencia fundamental de la señal
Tp = 1/Fo # Periodo de la señal
Fs = 1000 # Frecuencia de muestreo, mayor a 120 para que cumpla con el teorema de Nyquist
T = 1/Fs # Periodo de muestreo
t = np.arange(0, Tp+T, T)

senal = 4*np.sin(40*np.pi*t + (np.pi/4))+ 16*np.cos(80*np.pi*t)+ 5


plt.plot(t,senal)
plt.title("Señal senoidal")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.show()
```
